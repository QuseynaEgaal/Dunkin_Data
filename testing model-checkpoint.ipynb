# Import packages and dataset ▶️

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Split the data ✂️

```python
X = df['PTS']

y = df['WS']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=.15)
```

X: Winshares are our feature variables

Y: Points per game are our target variables

- We split the data into a training set and a testing set. 

- Visualise these sets separately

```python
plt.scatter(X_train, y_train, alpha=0.7, label='Training data', color='r')
plt.scatter(X_test, y_test, alpha=0.7, label='Testing data', color='g')
plt.legend()
```

# Train & fit the model 🤖

1. We fit the model to the training dataset

2. Make the predictions using the testing dataset

3. Plot the predictive model and the actual testing values

```python
linr = LinearRegression()
```

```python
linr.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
y_pred = linr.predict(X_test.values.reshape(-1,1))

plt.plot(X_test, y_pred, label='Linear regression', color='r')
plt.scatter(X_test, y_test, label='Actual data', color='g')
plt.legend()

plt.savefig('linr', dpi=300, facecolor='white',bbox_inches='tight' )
```

# Make a prediction & calculate error 🧠

```python
linr.predict(np.array([[150]]))[0]
```

```python
from sklearn.metrics import mean_squared_error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```
